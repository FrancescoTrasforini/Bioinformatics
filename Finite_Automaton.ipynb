```python
import numpy as np

def transition_function(pattern,alphabet):
  print("Pattern: ",pattern)
  pattern_length = len(pattern)
  gamma = []
  suffix = []
  # current_state is also the current longest prefix of "pattern" which is also
  # a suffix of the already read text
  for current_state in range (0,pattern_length):
    print("Current State: ", current_state)
     
    for letter in alphabet:
      print("Letter in consideration: " + letter)
      arr_state = min(pattern_length+1,current_state+2)
      print("Min arr_state: ", arr_state)
      executed = False
      #arr_state = arr_state - 1
      #suffix = pattern[:current_state]
      #suffix.append(letter)
      #print("suffix: ", suffix)
      #prefix = pattern[:arr_state-1]
      #print("prefix: ", prefix)
      i = 0 #loop count
      
      while(arr_state > 0 and (current_state-i) >= 0):
        
        arr_state = arr_state - 1
        suffix = pattern[i:current_state]
        suffix.append(letter)
        print("suffix: ", suffix)
        prefix = pattern[:arr_state]
        print("prefix: ", prefix)
        if(prefix == suffix):
          executed =True
          break
        i += 1

      if(executed == False):
        arr_state = 0
      transition = [current_state,letter,arr_state]  
      print("Transition: ",transition)
      gamma.append(transition)
  # transitions from the "final" state should be handled in a special way
  # I avoid managing them inside the previous loop to reduce the amount of
  # unnecessary checks
  current_state = pattern_length
  print("Current State: ", current_state) 
  for letter in alphabet:
    print("Letter in consideration: " + letter)
    arr_state = min(pattern_length+1,current_state+2)
    print("Min arr_state: ", arr_state) 
    i = 0 #loop count 
    while(arr_state > 0 and (current_state-i) >= 0):
      arr_state = arr_state - 1
      suffix = pattern[i+1:current_state]
      suffix.append(letter)
      print("suffix: ", suffix)
      prefix = pattern[:arr_state]
      print("prefix: ", prefix)
      if(prefix == suffix):
        # it will always execute at least once, no need to check if execution
        # happened. Just break out of the while loop if condition is met
        break
      i += 1
    transition = [current_state,letter,arr_state]  
    print("Transition: ",transition)
    gamma.append(transition)
    print("Gamma :", gamma)
  return gamma

        
```

```python
P = ["a","t","a","t"]
Alphabet = ["a","t","c"]
trans_function = transition_function(P,Alphabet)
print("Transition Function values: ", trans_function)
```
